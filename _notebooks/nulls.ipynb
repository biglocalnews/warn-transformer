# Nulls

Exploring empty cells in the dataset.

```python
import pandas as pd
```

```python
df = pd.read_csv("~/.warn-transformer/processed/consolidated.csv")
```

```python
df.info()
```

```python
df.head()
```

## Dates

```python
df[df.date.isnull()].groupby("postal_code").size().rename("total").reset_index()
```

## Company

```python
df[df.company.isnull()].groupby("postal_code").size().rename("total").reset_index()
```

## Location

```python
df[df.location.isnull()].groupby("postal_code").size().rename("total").reset_index()
```

## Jobs

```python
df[df.jobs.isnull()].groupby("postal_code").size().rename("total").reset_index()
```
